# Using the 2008 year ressetion from 2006 to 2011, check the infletion rate from 2019 to 2022 with a consistent time frame.

```python
#inflation during  2006 to 2011 ( 5 year)
first_df = avg_df.iloc[93:98]
first_df_row = first_df.to_numpy().flatten()
first_df_row = first_df_row[0:18]
plt.figure(figsize=(15, 8))
columns = [i for i in range(18)]
plt.xticks(range(0, 18, 1))
plt.yticks(range(int(min(first_df_row)), int(max(first_df_row)), 2))
plt.scatter(columns, first_df_row)
plt.xlabel('Month #')
```

* The up and down rate is measured every ten months.

```python
#inflation during 2018 to 2022 ( 5 year)
second_df = avg_df.iloc[102:107]
second_df_row = second_df.to_numpy().flatten()
second_df_row = second_df_row[0:48]
plt.figure(figsize=(15, 8))
columns = [i for i in range(48)]
plt.xticks(range(0, 48, 4))
plt.yticks(range(int(min(second_df_row)), int(max(second_df_row)), 2))
plt.scatter(columns, second_df_row)
plt.xlabel('Month #')
```

# conclusion

During four 12-month intervals, the rate of increase and reduction is measured. Don't you think we're heading for a high infletion rate?
